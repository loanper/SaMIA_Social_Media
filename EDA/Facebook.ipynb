### Analysis the lenght of each responses

```python
data['comment_length_words'] = data['comment_text'].astype(str).apply(lambda x: len(x.split()))
data['comment_length_chars'] = data['comment_text'].astype(str).apply(len)

print("Words :")
print(data['comment_length_words'].describe())
print("Characters :")
print(data['comment_length_chars'].describe())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Histogramme de toutes les longueurs
axes[0, 0].hist(data['comment_length_words'], bins=50, edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('Longueur (mots)')
axes[0, 0].set_ylabel('Fréquence')
axes[0, 0].set_title('Distribution des longueurs (Tout)')
axes[0, 0].axvline(data['comment_length_words'].mean(), color='red', linestyle='--', label=f'Moyenne: {data["comment_length_words"].mean():.1f}')
axes[0, 0].axvline(data['comment_length_words'].median(), color='green', linestyle='--', label=f'Médiane: {data["comment_length_words"].median():.1f}')
axes[0, 0].legend()

# Histogramme zoomé (0-200 mots)
data_filtered = data[data['comment_length_words'] <= 200]
axes[1, 0].hist(data_filtered['comment_length_words'], bins=50, edgecolor='black', alpha=0.7, color='orange')
axes[1, 0].set_xlabel('Longueur (mots)')
axes[1, 0].set_ylabel('Fréquence')
axes[1, 0].set_title('Distribution des longueurs (0-200 mots)')
axes[1, 0].axvline(32, color='blue', linestyle='--', label='32 tokens (SaMIA)')
axes[1, 0].axvline(64, color='purple', linestyle='--', label='64 tokens (SaMIA)')
axes[1, 0].axvline(128, color='brown', linestyle='--', label='128 tokens (SaMIA)')
axes[1, 0].legend()

plt.tight_layout()
plt.show()
```

### Percentage of words within Samia thresholds

```python

thresholds = [32, 64, 128]
for threshold in thresholds:
    count_below = (data['comment_length_words'] < threshold).sum()
    count_above = (data['comment_length_words'] >= threshold).sum()
    pct_below = (count_below / len(data)) * 100
    pct_above = (count_above / len(data)) * 100
    
    print(f"Seuil: {threshold} mots")
    print(f"  - Commentaires < {threshold} mots: {count_below} ({pct_below:.1f}%)")
    print(f"  - Commentaires >= {threshold} mots: {count_above} ({pct_above:.1f}%)")
    print()
```

## Analysisis of users et comments by users

Since the median is 9, we are looking to merge comments from different users.

```python
n_unique_users_by_name = data['user_name'].nunique()
n_unique_users_by_id = data['user_id'].nunique()

print(f"Nombre d'utilisateurs uniques (par user_name): {n_unique_users_by_name}")
print(f"Nombre d'utilisateurs uniques (par user_id): {n_unique_users_by_id}")
print(f"Nombre total de commentaires: {len(data)}")

# Commentaires par utilisateur
comments_per_user = data.groupby('user_id').size()
print(f"\nUser with only one comment: {(comments_per_user == 1).sum()} ({(comments_per_user == 1).sum() / len(comments_per_user) * 100:.1f}%)")
print(f"User with 2-5 comments: {((comments_per_user >= 2) & (comments_per_user <= 5)).sum()}")
print(f"User with 6-10 comments: {((comments_per_user >= 6) & (comments_per_user <= 10)).sum()}")
print(f"User with more than 20 comments: {(comments_per_user > 10).sum()}")
```

```python
# Visualisation de la distribution des commentaires par utilisateur
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Histogramme
axes[0].hist(comments_per_user, bins=range(1, comments_per_user.max() + 2), edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Nombre de commentaires par utilisateur')
axes[0].set_ylabel('Nombre d\'utilisateurs')
axes[0].set_title('Distribution du nombre de commentaires par utilisateur')
axes[0].set_xticks(range(1, comments_per_user.max() + 1))

# Top 20 utilisateurs
top_users = comments_per_user.nlargest(20)
axes[1].barh(range(len(top_users)), top_users.values)
axes[1].set_xlabel('Nombre de commentaires')
axes[1].set_ylabel('Utilisateur (rang)')
axes[1].set_title('Top 20 utilisateurs par nombre de commentaires')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()

print(f"\nMax commentaires par utilisateur: {comments_per_user.max()}")
```

## fusion of comments by user



```python
# Fusionner les commentaires par utilisateur
merged_comments = data.groupby('user_id').agg({
    'comment_text': lambda x: ' '.join(x.astype(str)),
    'user_name': 'first',
    'comment_length_words': 'sum'  # Somme des longueurs
}).reset_index()

# Recalculer la longueur après fusion
merged_comments['merged_length_words'] = merged_comments['comment_text'].apply(lambda x: len(str(x).split()))


print(f"Number of user before fusion: {len(merged_comments)}")
print(f"Number of user after fusion: {len(data)}")

print("\n=== STATISTICS OF LENGHTS AFTER FUSION ===")


print("\n=== QUARTILES APRÈS FUSION ===")
print(f"25% of comments are <= {merged_comments['merged_length_words'].quantile(0.25):.0f} mots")
print(f"50% of comments are <= {merged_comments['merged_length_words'].quantile(0.50):.0f} mots (median)")
print(f"75% of comments are <= {merged_comments['merged_length_words'].quantile(0.75):.0f} mots")
print(f"90% of comments are <= {merged_comments['merged_length_words'].quantile(0.90):.0f} mots")

```

```python
# Comparaison visuelle : avant et après fusion
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Distribution avant fusion (zoomée)
axes[0, 0].hist(data[data['comment_length_words'] <= 200]['comment_length_words'], 
                bins=50, edgecolor='black', alpha=0.7, color='skyblue')
axes[0, 0].set_xlabel('Longueur (mots)')
axes[0, 0].set_ylabel('Fréquence')
axes[0, 0].set_title('Distribution BEFORE fusion (0-200 mots)')
axes[0, 0].axvline(32, color='blue', linestyle='--', label='32 tokens')
axes[0, 0].axvline(64, color='purple', linestyle='--', label='64 tokens')
axes[0, 0].axvline(128, color='brown', linestyle='--', label='128 tokens')
axes[0, 0].axvline(data['comment_length_words'].median(), color='red', linestyle='-', linewidth=2, label=f'Médiane: {data["comment_length_words"].median():.0f}')
axes[0, 0].legend()

# Distribution après fusion (zoomée)
axes[0, 1].hist(merged_comments[merged_comments['merged_length_words'] <= 200]['merged_length_words'], 
                bins=50, edgecolor='black', alpha=0.7, color='lightcoral')
axes[0, 1].set_xlabel('Longueur (mots)')
axes[0, 1].set_ylabel('Fréquence')
axes[0, 1].set_title('Distribution AFTER fusion (0-200 mots)')
axes[0, 1].axvline(32, color='blue', linestyle='--', label='32 tokens')
axes[0, 1].axvline(64, color='purple', linestyle='--', label='64 tokens')
axes[0, 1].axvline(128, color='brown', linestyle='--', label='128 tokens')
axes[0, 1].axvline(merged_comments['merged_length_words'].median(), color='red', linestyle='-', linewidth=2, label=f'Médiane: {merged_comments["merged_length_words"].median():.0f}')
axes[0, 1].legend()

# Comparaison des distributions cumulatives
sorted_before = np.sort(data['comment_length_words'])
cumulative_before = np.arange(1, len(sorted_before) + 1) / len(sorted_before) * 100
sorted_after = np.sort(merged_comments['merged_length_words'])
cumulative_after = np.arange(1, len(sorted_after) + 1) / len(sorted_after) * 100

axes[1, 0].plot(sorted_before, cumulative_before, linewidth=2, label='AVANT fusion', color='skyblue')
axes[1, 0].plot(sorted_after, cumulative_after, linewidth=2, label='APRÈS fusion', color='lightcoral')
axes[1, 0].axvline(32, color='blue', linestyle='--', alpha=0.5)
axes[1, 0].axvline(64, color='purple', linestyle='--', alpha=0.5)
axes[1, 0].axvline(128, color='brown', linestyle='--', alpha=0.5)
axes[1, 0].set_xlabel('Longueur (mots)')
axes[1, 0].set_ylabel('Pourcentage cumulatif (%)')
axes[1, 0].set_title('Comparaison of cumulative distributions')
axes[1, 0].set_xlim(0, 200)
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].legend()

# Comparaison en barres des seuils SaMIA
thresholds = [32, 64, 128]
before_pct = [(data['comment_length_words'] >= t).sum() / len(data) * 100 for t in thresholds]
after_pct = [(merged_comments['merged_length_words'] >= t).sum() / len(merged_comments) * 100 for t in thresholds]

x = np.arange(len(thresholds))
width = 0.35

axes[1, 1].bar(x - width/2, before_pct, width, label='BEFORE fusion', color='skyblue')
axes[1, 1].bar(x + width/2, after_pct, width, label='AFTER fusion', color='lightcoral')
axes[1, 1].set_xlabel('Threshold (words)')
axes[1, 1].set_ylabel('Perrcentage >= Threshold (%)')
axes[1, 1].set_title('Comparaison : % of texts >= Threshold SaMIA')
axes[1, 1].set_xticks(x)
axes[1, 1].set_xticklabels(thresholds)
axes[1, 1].legend()
axes[1, 1].grid(True, alpha=0.3, axis='y')

# Ajouter les valeurs sur les barres
for i, (b, a) in enumerate(zip(before_pct, after_pct)):
    axes[1, 1].text(i - width/2, b + 1, f'{b:.1f}%', ha='center', fontsize=9)
    axes[1, 1].text(i + width/2, a + 1, f'{a:.1f}%', ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

## 📊 RÉSUMÉ ET RECOMMANDATIONS

### Points clés de l'analyse :

#### 1. **Structure du dataset**
- **1000 commentaires** de **804 utilisateurs uniques**
- Colonnes principales : `user_name`, `user_id`, `comment_text`, `date_created`, etc.

#### 2. **Distribution des longueurs de commentaires (AVANT fusion)**
- **Moyenne** : 29.4 mots
- **Médiane** : 9 mots ⚠️ (très court!)
- **75% des commentaires** ont ≤ 28 mots
- **77.3%** des commentaires ont < 32 mots
- **88.6%** des commentaires ont < 64 mots
- **95.5%** des commentaires ont < 128 mots

#### 3. **Analyse des utilisateurs**
- **87.1%** des utilisateurs n'ont qu'**1 seul commentaire**
- Max : 9 commentaires par utilisateur
- Moyenne : 1.24 commentaires par utilisateur

#### 4. **Distribution après FUSION par utilisateur**
- **804 textes** (au lieu de 1000)
- **Médiane** : toujours 9 mots (pas de grand changement)
- **22.1%** des utilisateurs ont ≥ 32 mots (vs 22.7% avant)
- **12.9%** des utilisateurs ont ≥ 64 mots (vs 11.4% avant)
- **5.3%** des utilisateurs ont ≥ 128 mots (vs 4.5% avant)

---

### 🎯 RECOMMANDATIONS pour SaMIA

#### **Longueur de texte à utiliser : 32 tokens**

**Justification :**
1. ✅ **~22-23%** du dataset a des textes ≥ 32 mots (suffisant pour l'analyse)
2. ✅ Les textes de 64 et 128 tokens sont **trop rares** (11-13% et 4-5% seulement)
3. ✅ La médiane de **9 mots** indique que les commentaires Facebook sont **courts par nature**
4. ✅ 32 tokens est un bon compromis pour capturer le contenu significatif des commentaires de réseaux sociaux

#### **Stratégie de fusion : OPTIONNELLE mais PEU UTILE**

**Observation :**
- La fusion par utilisateur **ne change presque rien** car 87% des utilisateurs n'ont qu'1 commentaire
- Gain marginal : +0.6% pour 32 tokens, +1.5% pour 64 tokens, +0.8% pour 128 tokens
- **Pas de fusion recommandée** : garder les commentaires individuels pour préserver le contexte

---

### 📋 Actions suggérées :

1. **Filtrer les données** : Garder uniquement les commentaires ≥ 32 mots
2. **Alternative** : Si besoin de plus de données, considérer les textes de 20-32 mots
3. **Ne pas fusionner** les commentaires par utilisateur (gain minimal)
4. **Tokenization** : Utiliser un tokenizer approprié (ex: BERT, GPT) pour compter précisément les tokens
